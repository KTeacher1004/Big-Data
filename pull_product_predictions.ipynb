# Pull product_predictions collection
Fetch MongoDB 'product_predictions' collection from 'Assignment3' DB and save locally.

````python
# Install deps if needed (uncomment next line)
# !pip install pymongo pandas
import os
from pymongo import MongoClient
import pandas as pd

MONGODB_URI = os.getenv('MONGODB_URI','mongodb+srv://admin:adminpassword@assignment3.dhfn7vh.mongodb.net/?retryWrites=true&w=majority&appName=Assignment3')
DB_NAME='Assignment3'
COLLECTION='product_predictions'

client = MongoClient(MONGODB_URI)
db = client[DB_NAME]
coll = db[COLLECTION]
docs = list(coll.find({}))
print(f'Fetched {len(docs)} documents')
````

````python
from bson import ObjectId

for d in docs:
    if isinstance(d.get('_id'), ObjectId):
        d['_id'] = str(d['_id'])

df = pd.DataFrame(docs)
print(df.shape)
df.head()
````

````python
output_csv='product_predictions.csv'
output_json='product_predictions.json'
df.to_csv(output_csv, index=False)
df.to_json(output_json, orient='records')
print('Saved', output_csv, 'and', output_json)
````
